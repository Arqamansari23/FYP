```python
from langchain.document_loaders import UnstructuredURLLoader
```

```python
loader = UnstructuredURLLoader(
    urls = [
        "https://www.moneycontrol.com/india/stockpricequote/auto-lcvshcvs/tatamotors/TM03",
        # "https://www.moneycontrol.com/news/business/markets/market-corrects-post-rbi-ups-inflation-forecast-icrr-bet-on-these-top-10-rate-sensitive-stocks-ideas-11142611.html"
    ]
)
```

```python
data = loader.load()
data
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=200)
documents=text_splitter.split_documents(data)
```

```python
a=[]
for i in range (len(documents)):
    if len(documents[i].page_content)>=500:
        a.append(i)
print(a)
```

```python
from langchain.embeddings import OllamaEmbeddings
from langchain.vectorstores import FAISS

# Initialize Ollama's LLaMA embeddings
embeddings = OllamaEmbeddings(model="llama3.1")  # Replace with the version you are using, e.g., "llama-3.1"
```

```python
db = FAISS.from_documents(documents[:15], embeddings)  
```

```python
query = "what is the latest price of stock of tata moters  ?"
results = db.similarity_search(query, k=3)  # Get the top 3 matching documents

# Print the results
for idx, result in enumerate(results, 1):
    print(f"Result {idx}:")
    print(result.page_content)
    print("-" * 50)
```

```python
from langchain_community.llms import Ollama
## Load Ollama LAMA2 LLM model
llm=Ollama(model="llama3.1")
llm
```

```python
# Generate the context from search results
context = "\n".join([result.page_content for result in results])
```

```python
# Generate the context from search results
context = "\n".join([result.page_content for result in results])

# Define the query (example: user asking about a company's stock price)
user_query = "What is the latest price of the Tata moters  company's stock?"

# Final prompt for LLaMA with generalized context and role
prompt = f"""
I am a Financial Equity Research Analyst with expertise in analyzing companies and guiding investment decisions.
I provide insights into stock prices, company fundamentals, and news sentiment analysis to help investors make informed decisions.

Context:
{context}

User Query: {user_query}

Based on the context provided, analyze the current stock price and provide a sentiment analysis of any relevant news surrounding this company.
Additionally, suggest whether this stock is a good investment opportunity at this time.
"""
```

```python
response = llm.generate([prompt])
print(response)
```
